```python
def no_null_objects(data, columns=None):
    """
    Selects rows with no NaN values in the specified columns.
    If no columns are specified, it checks all columns.
    """
    if columns is None:
        columns = data.columns
    # Use logical NOT to filter rows without any NaN values in the specified columns
    return data[np.logical_not(np.any(data[columns].isnull().values, axis=1))]

def splitDataFrameList(df, target_column, separator):
    """
    Splits the values in the target column of a DataFrame into separate rows based on a separator.
    Other column values are duplicated for the new rows.

    Parameters:
    - df: DataFrame to split
    - target_column: The column containing the values to split
    - separator: The symbol used to perform the split

    Returns:
    - A new DataFrame with the split values in separate rows.
    """
    def splitListToRows(row, row_accumulator, target_column, separator):
        # Split the target column value using the separator
        split_row = row[target_column].split(separator)
        for s in split_row:
            # Create a new row for each split value
            new_row = row.to_dict()
            new_row[target_column] = s
            row_accumulator.append(new_row)

    # Initialize a list to store the new rows
    new_rows = []
    # Apply the splitting function to each row in the DataFrame
    df.apply(splitListToRows, axis=1, args=(new_rows, target_column, separator))
    # Create a new DataFrame from the accumulated rows
    new_df = pd.DataFrame(new_rows)
    return new_df
```

## Heat map over unique countries

```python
# Filter rows with no null values in the 'countries_en' column and split the 'countries_en' column into separate rows based on the ',' separator
food_countries = splitDataFrameList(no_null_objects(food, ["countries_en"]), "countries_en", ",")

# Count the occurrences of each unique value in the 'countries_en' column and store the result in the 'countries' variable
countries = food_countries["countries_en"].value_counts()
```

# Correlations

```python
# Calculate the correlation of 'nutrition-score-fr_100g' with other numerical columns
# Select only numeric columns to avoid errors during correlation calculation
numeric_food = food.select_dtypes(include=[np.number])

# Calculate the correlation of 'nutrition-score-fr_100g' with other numerical columns
correlations = numeric_food.corr()['nutrition-score-fr_100g'].sort_values(ascending=False)

# Display the top 10 variables with the highest positive correlation
print("Top 10 positive correlations:")
display(correlations.head(10))

# Display the top 10 variables with the highest negative correlation
print("\nTop 10 negative correlations:")
display(correlations.tail(10))
```

```python
# Calculate the Pearson correlation matrix for the numeric columns in the DataFrame
corr = numeric_food.corr(method="pearson")

# Create a matplotlib figure and axes with a specified size
f, ax = plt.subplots(figsize=(15, 15))

# Plot a heatmap of the correlation matrix using seaborn
# Mask the upper triangle of the heatmap to avoid redundancy
# Use a diverging color palette for better visualization
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), 
            cmap=sns.diverging_palette(220, 10, as_cmap=True), square=True, ax=ax)
```
